# datatype conversions

```python
# value column concersion after removing the necessary symbols and letters
```

```python
fifa['Value']=fifa['Value'].str.replace('€','')
fifa['Value']=fifa['Value'].str.replace('K','000')
fifa['Value']=fifa['Value'].str.replace('M','000000')
```

```python
fifa.dtypes
```

```python
# value column from object to float conversion
fifa['Value'] = fifa['Value'].astype('float')
```

```python
fifa.dtypes
```

```python
# 'Wage' conversion to float
```

```python
fifa['Wage']=fifa['Wage'].str.replace('€','')
fifa['Wage']=fifa['Wage'].str.replace('K','000')
fifa['Wage']=fifa['Wage'].str.replace('M','000000')
```

```python
fifa['Wage'] = fifa['Wage'].astype('float')
```

```python
fifa['Wage'].dtype
```

```python
# 'Joined' conversion
```

```python
fifa['Joined']=pd.DatetimeIndex(fifa['Joined']).year
fifa['Joined']
```

```python
fifa['Joined']= fifa['Joined'].astype('float')
```

```python
fifa['Joined'].dtype
```

```python
# 'Contract Valid Until' conversion from string to float
```

```python
fifa['Contract Valid Until'].dtype
```

```python
fifa['Contract Valid Until'] = pd.to_datetime(fifa['Contract Valid Until'])
```

```python
fifa['Contract Valid Until']
```

```python
# 'Height' conversion to float by replacing ' with .'
```

```python
fifa['Height'].dtype
```

```python
fifa['Height']
```

```python
fifa['Height']=fifa['Height'].str.replace("'",".")
```

```python
fifa['Height']=fifa['Height'].astype("float")
print(fifa['Height'].dtype)
fifa['Height']
```

```python
# weight conversion with required changes
fifa['Weight']=fifa['Weight'].str.replace("lbs","")
fifa['Weight']=fifa['Weight'].astype("float")
```

```python
fifa['Weight'].dtype
```

```python
# release cause conversion
fifa['Release Clause']
```

```python
fifa['Release Clause']=fifa['Release Clause'].str.replace('€','')
fifa['Release Clause']=fifa['Release Clause'].str.replace('K','1000')
fifa['Release Clause']=fifa['Release Clause'].str.replace('M','1000000')
fifa['Release Clause']=fifa['Release Clause'].astype("float")
```

```python
print(fifa['Release Clause'].dtype)
fifa['Release Clause']
```

# 4.	Check for missing values and do a mean imputation where necessary.

```python
fifa.isnull().sum()/len(fifa) * 100
```

```python
# release_cause, joined,contract_valid_untill have numerical missing values in range of 1-10 % 
# replacing missing values wouldnt change the distribution as percent is less
# yet we are proceeding with missing values imputation using mean imputation
```

```python
fifa['Release Clause'].describe()
# since the percentage is less there qouldnt be any effect of imputation on the distribution which can be observed from below plots
```

```python
sns.distplot(fifa['Release Clause'])
```

```python
fifa['Release Clause'].fillna(fifa['Release Clause'].mean(),inplace = True)
```

```python
sns.distplot(fifa['Release Clause'])
```

```python
sns.distplot(fifa['Joined'])
```

```python
print(round(fifa['Joined'].mean()))
print(fifa['Joined'].median())
# there isnt much difference between mean and median hence replacing with mean wouldnt effect the distribution
```

```python
fifa['Joined'].fillna(round(fifa['Joined'].mean()),inplace = True)
```

```python
sns.distplot(fifa['Joined'])
```
